```python
data = original_data.copy()
data.Age[data.Age.isnull()] = data.Age.mean()
data
```

# Baseline

```python
X = data[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
y = data['Survived']
```

```python
X, y
```

```python
y.value_counts().plot.pie(figsize=(2, 2))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
y_test.mean(), y_train.mean()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train), model.score(X_test, y_test)
```

### lets try to overfit

```python
def fast_score(model, X, y, y_col='Survived'):
    y = data[y_col];
    X = data.drop(y_col, axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
    model.fit(X_train, y_train)
    print('train set score', model.score(X_train, y_train), 'test set score', model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
train_data = data[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]
train_data
```

```python
fast_score(LogisticRegression(), train_data, 'Survived')
```

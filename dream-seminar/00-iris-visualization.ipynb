```python
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
iris.feature_names, iris.target_names
```

```python
dim0 = 1
dim1 = 2

plt.figure(figsize=(10, 8))

for target in range(len(iris.target_names)):
    plt.scatter(
        x=iris.data[:, dim0][iris.target == target],
        y=iris.data[:, dim1][iris.target == target],
        label=iris.target_names[target]
    )

plt.xlabel(iris.feature_names[dim0], fontsize=18)
plt.ylabel(iris.feature_names[dim1], fontsize=18)

plt.legend(fontsize=15)
plt.title('Iris Data', fontsize=20)
```

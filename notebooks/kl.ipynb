```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as pl
```

```python
def P(x):
    sigma = 0.5
    l = norm(-3, 1).pdf(x)
    r = norm(3, 1).pdf(x)
    return l * sigma + r * (1 - sigma)
```

```python
def Q(x, mu, sig):
    return norm(mu, sig).pdf(x)
```
